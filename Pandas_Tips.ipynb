**Splitting a String into Columns**

```python
import pandas as pd
df=pd.DataFrame({'name':['Arbaz Khan','Ahmed Ai','Akhttar Shah'],'Loc':['USA','Germany','Itally']})
df
```

```python
df.name.str.split(' ',expand=True) #Split the string into two columns. Expand=True zroor lagan hy. Or in ' ' js se separae krna ho wo .
df[['F_N','L_N']]=df.name.str.split(' ',expand=True) # Ab is mei df name column ko split kr diya hai. Aur is mei do ne columns bna dei hy splited columns.
df
```

```python
# Refininig DAta Sets
df=df[['F_N','L_N']] #Yene purany data sets mei sy selected data sets ko bna lia. 
df
```

**Aggregating by Multiple Groups/Functions**

```python
import pandas as pd
import seaborn as sns
d=sns.load_dataset('Titanic')
d.head()
```

```python
d.groupby('sex').count() # Ab is mei groupby sex kr dia yene k sirf male or female honge aur count se pta laga k hnn kitny kitny.
len(d.groupby('sex')) # Is Len function se idea hota hy  k us columns mei total variety kitni honge. Isme 2 honge male aur female. Isliy
d.groupby(['sex','pclass','class']).count()
```

**Selecting a  Specific Column or Rows** 

```python
d[['age','who']].head()
d.loc[2] # Is trah specific row ko select kry hnn . ioc bhe use hota hy us ko bhe dekh lena.
d.describe()
d['fare'].unique() # Iss trah kse collumn mei se koi bhe unique value nikalny k iye. 
d.loc[d['fare']==53.1000] # Aur is traika se kse bhe specific value k corresponding row nikalte hy.
```

```python
d.describe().loc["min":'max'] # Iss trah data set mei se rows ko limit laga k nikal liya hy.

d.describe().loc["min":'max','age'] # Iss traika se column a jae gaa. 

d.describe().loc["min":'max',['age','fare']] # Rows k sath multiple columns k liye ye [] brackets lagany hy.

d.describe().loc["min":'max','age':'fare'] # Iss traika sy coumns ko imi de sky hnn.
```
